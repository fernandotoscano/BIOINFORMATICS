## **Preparing the X and Y Data Matrices**

### **X data matrix**

**AND SO WE'RE GONNA READ THE DESCRIPTORS OUTPUT INTO "df3_X", AND SO HERE WE'RE GOING TO PREPARE THE X, AND "Y" DATA MATRICES. AND THE "X" DATA WILL COMPRICE OF THE MOLECULAR DESCRIPTORS WHICH ARE THE PUBCHEM FINGRERPRINTS SO LET'S HAVE A LOOK:**

```
df3_X = pd.read_csv('descriptors_output.csv')
```

```
df3_X
```

**AND SO WE'RE GOING TO DELETE THE 1ST COLUMN HERE THE "NAME" BECAUSE WE WANT ONLY THE MOLECULAR FEATURES SO LET'S DROP IT USING THE DOT DROP FUNCTION (df3_X = df3_X.drop(colummns=['Name'])", AND THE NAME OF THE COLUMN THAT WE WANT IT TO BE DROPED AND THEN WE CAN SEE THAT THE "Name" COLUMN HAS BEEN DROPPED (AND THEN WE REASSIGN IT BACK TO "df3_X" ): **

```
df3_X = df3_X.drop(columns=['Name'])
df3_X
```

## **Y variable**

[59:14] **SO NOW LET'S CREATE THE "Y" DATA MATRIX (THE DEPENDENT VARIABLE), AND HERE WE'RE GOING TO TAKE THE "pIC50" COLUMN DIRECTLY FROM THE "df3" DATAFRAME WHICH IS THE INITIAL LOADED DATA FRAME AND THEN WE'RE GOING TO ASSIGN IT TO THE "df3_Y":**

### **Convert IC50 to pIC50**

```
df3_Y = df3['pIC50']
df3_Y
```

## **Combining X and Y variable**

**AND THEN HERE WE'RE GOING TO COMBINE "X" AND "Y" TOGETHER SO THIS REALLY DEPENDS BUT FOR PORTABILITY WE'RE GONNA COMBINEIT AND WE'RE GONNA OUTUP IT INTO A ".csv" FILE WHICH WE WILL BE UPLOADING TO GITHUB, AND THEN WE'RE GONNA USE THAT FOR PART_4:**

```
dataset3 = pd.concat([df3_X,df3_Y], axis=1)
dataset3
```
